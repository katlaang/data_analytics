# THE DATA

This project is based on data available on the open database Kaggle: Video game sales and Ratings. The data set has a list of video game sales in various regions along with their corresponding critic and user ratings. A list of variables included in the data are given below:
1.	Name - The game's name
2.	Platform - Platform of the games release
3.	Year_of_Release - Year of the game's release
4.	Genre - Genre of the game
5.	Publisher - Publisher of the game
6.	NA_Sales - Sales in North America (in millions)
7.	EU_Sales - Sales in Europe (in millions)
8.	JP_Sales - Sales in Japan (in millions)
9.	Other_Sales - Sales in the rest of the world (in millions)
10.	Global_Sales - Total worldwide sales (in millions)
11.	Critic_score - Aggregate score compiled by Metacritic staff
12.	Critic_count - The number of critics used in coming up with the critic score
13.	User_score - Score by Metacritic's subscribers
14.	User_count - Number of users who gave the user score
15.	Rating - The ESRB ratings

The data matrix has 17416 rows and 15 rows.

# DATA CLEANING 
In order to have accurate result from the data. Firstly, we have to clean abnormal data by following steps:
1. Screen the data  
2. Find the missing values and abnormal values.
3. Try to remove or adjust missing values and abnormal values

In this study, we plan to analyze video games status in different years, therefore we plan to drop all the rows containing missing value of Year_of_Release. New datafram will be created for further analysis.For other columns i.e. critics_scores, user_scores, it is hard to do any fill up action, and it will not affect the main focus of our analysis, we plan to keep this missing values.
In column of rating, we replace "K-A" to "E". In ESRB rating system, K-A was the previous name of E, since 1998, the K-A was not used anymore.

2. The third generation consoles (1983â€“1995) (8-bit)
  
  At the end of 1983, several factors, including a market flooded with poor-quality games, the commercial failure of several important Atari 2600 titles, and home computers emerging as a new and more advanced gaming platform, caused the industry to experience a severe downturn. This was the "crash" of the video game industry. It bankrupted several companies that produced North American consoles and games from late 1983 to early 1984. Whilst a broken gaming industry in the US took several local businesses to bankruptcy and practically ended retail interest in video gaming products, an 8-bit third generation of video game consoles started in Japan as early as 1983 with the release of both Nintendo's Family Computer ("Famicom") and Sega's SG-1000 on July 15. The first clearly trumped the second in terms of commercial success in the country, causing Sega to replace it, two years later, by a severely improved and modernized version called the Sega Mark III. From the table we can see comparative game market shares between North-America and Japan. In addition, the top five games publishers were from Japan companies.
